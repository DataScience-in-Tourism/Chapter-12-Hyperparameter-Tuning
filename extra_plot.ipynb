```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame({'Overall Loss (MSE)': [97.775, 96.853, 95.251, 18.175, 0.00214, 7.240], 'Hyperparameter-tuning technique': ['Default Parameters', 'Manual Search',
                                                                                                                                'Grid Search', 'Random Search',
                                                                                                                                'Bayesian Optimization', 'Genetic Algorithms']})
```

```python
table = pd.pivot_table(pd.DataFrame(data),
    values='Overall Loss (MSE)', index='Hyperparameter-tuning technique')
     
sns.heatmap(table)
```

```python
plt.figure(num=None, figsize=(10, 7), dpi=80, facecolor='w', edgecolor='k')
plt.plot(data['Overall Loss (MSE)'])
plt.xticks([i for i in range(0, len(data['Overall Loss (MSE)']))], data['Hyperparameter-tuning technique'], rotation=30, fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Hyperparameter-tuning technique", fontsize=16)
plt.ylabel("Overall Loss (MSE)", fontsize=16)
plt.title("Overall Loss (MSE) vs Hyperparameter-tuning technique", fontsize=18)
plt.savefig('bar5.svg', format='svg', dpi=120, bbox_inches='tight')
```
